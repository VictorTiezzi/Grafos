#### DADOS

```python
from igraph import *
#import random

V = {
    0: 'AC',
    1: 'AL',
    2: 'AP',
    3: 'AM',
    4: 'BA',
    5: 'CE',
    6: 'DF',
    7: 'ES',
    8: 'GO',
    9: 'MA',
    10: 'MT',
    11: 'MS',
    12: 'MG',
    13: 'PA',
    14: 'PB',
    15: 'PR',
    16: 'PE',
    17: 'PI',
    18: 'RJ',
    19: 'RN',
    20: 'RS',
    21: 'RO',
    22: 'RR',
    23: 'SC',
    24: 'SP',
    25: 'SE',
    26: 'TO'}

E = {
    0: [3, 21],
    1: [4, 16, 25],
    2: [13],
    3: [0, 10, 13, 21, 22],
    4: [1, 7, 8, 12, 16, 17, 25, 26],
    5: [14, 16, 17, 19],
    6: [8, 12],
    7: [4, 12, 18],
    8: [4, 6, 10, 11, 12, 26],
    9: [13, 17, 26],
    10: [3, 8, 11, 13, 21, 26],
    11: [8, 10, 12, 15, 24],
    12: [4, 6, 7, 8, 11, 18, 24],
    13: [2, 3, 9, 10, 22, 26],
    14: [5, 16, 19],
    15: [11, 23, 24],
    16: [4, 5, 14, 17],
    17: [4, 5, 9, 16, 26],
    18: [7, 12, 24],
    19: [5, 14],
    20: [23],
    21: [0, 3, 10],
    22: [3, 13],
    23: [15, 20],
    24: [11, 12, 15, 18],
    25: [1, 4],
    26: [4, 8, 9, 10, 13, 17]}


#random.shuffle(V)

g = Graph()
```

#### TRATAMENTO DOS VÃ‰RTICES

```python

tags_dos_verteces=[]

for x in V:
    tags_dos_verteces.append(V[x])

g.add_vertices(len(V))      #numeros de Verteces
```

#### TRATAMENTO DAS ARESTAS

```python

arestas=[]

for x in E:
    for y in E[x]:
        if not((y,x) in arestas):
            arestas.append((x,y))

g.add_edges(arestas)   #arestas
```

#### TRATAMENTO DAS CORES

```python
cores=['aqua', 'greenyellow', 'yellow', 'salmon', 'red']

Lista_ordenada = dict(sorted(E.items(), key=lambda x: len(x[1]), reverse=True))
cores_em_dicionario = {next(iter(Lista_ordenada)): cores[0]}

for item in Lista_ordenada.items():
    
    cor_existente = []
    for index in item[1]:    
        if index in cores_em_dicionario and cores_em_dicionario[index] not in cor_existente:
            cor_existente.append(cores_em_dicionario[index])
    
    for cor in cores:
        if cor not in cor_existente:
            cores_em_dicionario[item[0]] = cor
            break
            
cores_no_grafo = []                  
for cor in sorted(cores_em_dicionario.items()):
    cores_no_grafo.append(cor[1])
```

#### PLOTANDO O GRAFO

```python
plot(g,
        vertex_size=35,
        vertex_color=cores_no_grafo,
        vertex_label=tags_dos_verteces,
        edge_width=[1],
        edge_color=['black'],
        )
```
